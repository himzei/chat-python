```python
import pandas as pd 
df = pd.read_csv("type1_data1.csv")


# 1. "views" 컬럼의 결측 데이터를 0으로 대체하시오. 
# 2. "views" 컬럼에서 상위 10번째 값을 구하시오
# 3. "views" 컬럼에서 상위 10개의 값을 상위 10번째 값을 대체하시오. 
# 4. "views" 컬럼의 전체 합을 정수로 구하시오.
```

```python
df.head(1)
df.isnull().sum()
```

```python
# 1. 
df = df.fillna(0)
df.isnull().sum()
```

```python
# 2. "views" 컬럼에서 상위 10번째 값을 구하시오

df = df.sort_values("views", ascending=False) 
a = df[:10]
result2 = a.iloc[:10, -1].min()
result2
```

```python
# 3. "views" 컬럼에서 상위 10개의 값을 상위 10번째 값을 대체하시오. 


df.iloc[0:10, -1] = result2
df[:11]
```

```python
# 4. "views" 컬럼의 전체 합을 정수로 구하시오.

df["views"].sum()
```

```python
# 1. "views" 컬럼의 결측 데이터를 0으로 대체
df['views'] = df['views'].fillna(0)

# 2. "views" 컬럼에서 상위 10번째 값을 구하기
# 내림차순으로 정렬 후 10번째 값 (인덱스 9)
views_sorted = df['views'].sort_values(ascending=False)

# 데이터가 10개 이상인지 확인
if len(views_sorted) < 10:
    raise ValueError("데이터가 10개 미만입니다. 상위 10번째 값을 구할 수 없습니다.")

상위_10번째_값 = views_sorted.iloc[9]  # 인덱스는 0부터 시작하므로 9가 10번째
print(f"상위 10번째 값: {상위_10번째_값}")

# 3. "views" 컬럼에서 상위 10개의 값을 상위 10번째 값으로 대체
# 상위 10개 행의 인덱스를 찾아서 값을 대체
상위_10개_인덱스 = views_sorted.head(10).index
df.loc[상위_10개_인덱스, 'views'] = 상위_10번째_값

# 4. "views" 컬럼의 전체 합을 정수로 구하기
views_전체_합 = int(df['views'].sum())
print(f"views 컬럼 전체 합: {views_전체_합}")
```
