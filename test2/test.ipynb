```python
def calculate_average(scores):
    avg = round(sum(scores) / len(scores), 2) 

    if avg >= 90: 
        grade =  "A" 
    elif avg >= 80: 
        grade = "B" 
    elif avg >= 70: 
        grade = "C" 
    elif avg >= 60: 
        grade = "D" 
    else: 
        grade = "F" 

    return avg, grade


```

```python
name = "김철수"
avg, grade = calculate_average([85, 90, 88])
print(avg) 
print(grade)
```

```python
import pandas as pd
books = {
    "도서명" : ["데이터분석", "알고리즘", "자바입문", "웹개발", "파이썬"], 
    "재고" : [12, 20, 8, 3, 15], 
    "가격": [35000, 32000, 30000, 28000, 25000]
}

df = pd.DataFrame(books)
df
```

```python
df.isnull().sum()
```

```python
df = df.sort_values("가격", ascending=False)
```

```python
df["할인가격"] = (df["가격"] * (1 - 0.1)).astype(int)
df
```

```python
df.to_csv("books_data.csv", index=False)
```
