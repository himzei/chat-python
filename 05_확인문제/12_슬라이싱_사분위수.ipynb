```python
import pandas as pd 
df = pd.read_csv("type1_data1.csv")

#1. 데이터에서 결측치가 있는 데이터(행)를 삭제하시오. 
#2. 결측치가 제거된 데이터를 사용하여 앞에서부터 70%데이터를 구하시오.
#3. 앞에서 구한 70% 데이터 중 "views" 컬럼의 3사분위수에서 1사분위수 뺀 값을 정수로 구하시오 
```

```python
#1. 데이터에서 결측치가 있는 데이터(행)를 삭제하시오.
print("결측치 확인 ", df.isnull().sum())
df.dropna(inplace = True)
print("결측치 확인 ", df.isnull().sum())
```

```python
#2. 결측치가 제거된 데이터를 사용하여 앞에서부터 70%데이터를 구하시오.

df.shape
len_7 = len(df) * 0.7
len_7
```

```python
df.head()
```

```python
df = df.iloc[0:int(len_7)]
df.shape
```

```python
#3. 앞에서 구한 70% 데이터 중 "views" 컬럼의 3사분위수에서 1사분위수 뺀 값을 정수로 구하시오 

q_3 = df["views"].quantile(0.75)
print(q_3)

q_1 = df["views"].quantile(0.25) 
print(q_1)

result = int(q_3 - q_1)
result
```
