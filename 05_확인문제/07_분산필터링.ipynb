```python
import pandas as pd 
df = pd.read_csv("type1_data1.csv")

# 1. "f3" 컬럼이 gold 이면서 "f2" 컬럼이 2인 데이터를 찾으시오 
# 2. 찾은 데이터에서 "f1" 컬럼의 분산을 구하시오. (반올림 후 소수 둘째 자리까지 계산)
```

```python
df.head()
df.info()

```

```python
# 1. "f3" 컬럼이 gold 이면서 "f2" 컬럼이 2인 데이터를 찾으시오 
cond1 = df["f3"] == "gold"
cond2 = df["f2"] == 2 

result1 = df[cond1 & cond2]
```

```python
# 2. 찾은 데이터에서 "f1" 컬럼의 분산을 구하시오. (반올림 후 소수 둘째 자리까지 계산)

result2 = result1["f1"].var()
result2 = round(result2, 2)
result2 
```
