```python
import pandas as pd 

# 데이터 불러오기
df = pd.read_csv("type1_data1.csv")


# 1. "views" 컬럼의 1사분위수, 3사분위수, iqr구하시오
# (iqr = 3사분위수 - 1사분위수)
# 2. 이상치 조건에 맞는 데이터를 찾으시오 
# (이상치는 1사분위수 - (iqr *1.5)보다 작은 값고 3사분위 + (iqr * 1.5)보다 큰값)
# 3. 이상치 데이터의 "views" 컬럼 합을 정수로 구하시오 
```

```python
df.head(3)
```

```python
q3 = df["views"].quantile(0.75)
q1 = df["views"].quantile(0.25)
iqr = q3 - q1 
iqr
```

```python
# 2. 이상치 조건에 맞는 데이터를 찾으시오 
# (이상치는 1사분위수 - (iqr *1.5)보다 작은 값고 3사분위 + (iqr * 1.5)보다 큰값)

line1 = q1 - (iqr * 1.5)
line2 = q3 + (iqr * 1.5) 

cond1 = df["views"] < line1
cond2 = df["views"] > line2

df = df[cond1 | cond2]





```

```python
# 3. 이상치 데이터의 "views" 컬럼 합을 정수로 구하시오 
int(df["views"].sum())
```
