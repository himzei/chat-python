```python
import pandas as pd 

# 데이터 불러오기
df = pd.read_csv("type1_data1.csv")

# 1. "views" 컬럼의 표준편차를 구하시오 
# 2. "age" 컬럼의 이상치(소수점 나이, 음수 나이, 0살) 제거하고, "views"컬럼의 표준편차를 구하시오.
# 3. 이상치 제거 전후의 'views" 컬럼의 표준편차를 더하여, 반올림 후 소수 둘째 자리까지 구하시오. 
```

```python
# 1. "views" 컬럼의 표준편차를 구하시오 
result1 = df["views"].std()
```

```python
# 2. "age" 컬럼의 이상치(소수점 나이, 음수 나이, 0살) 제거하고, 
# "views"컬럼의 표준편차를 구하시오.

cond = df["age"] <= 0
df = df[~cond]

```

```python
df["age"].value_counts()
```

```python

cond = df["age"] == round(df["age"], 0)

# 7.0       ==    round(7.0, 0)
# 7.1       ==    round(7.1, 0)
# 8.0       ==    round(8.0, 0)
# 8.5       ==    round(8.5   , 0)

# 7.0 == 7 
# 7.1 == 7 
# 8.0 == 8
# 8.5 == 9

df = df[cond]
df["age"].value_counts()

```

```python
df["age"] == round(df["age"], 0)
```

```python
result2 = df["views"].std()
result2
```
