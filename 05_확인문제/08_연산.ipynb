```python
import pandas as pd 
df = pd.read_csv("type1_data1.csv")

# 1. 모든 나이(age)에 1을 더하시오. 
# 2. 20대의 "views" 평균과 30대의 "views" 평균의 차이의 절대값을 구하시오.
# (반올림 후 소수 둘째 자리까지 계산)
```

```python
# 1. 모든 나이(age)에 1을 더하시오. 
df["age"].describe()
df["age"] = df["age"] + 1
df["age"]
```

```python
# 2. 20대의 "views" 평균과 30대의 "views" 평균의 차이의 절대값을 구하시오.
# (반올림 후 소수 둘째 자리까지 계산)

df.head()
```

```python
# 20대의 데이터 

cond_20 = (df["age"] >= 20) & (df["age"] < 30)
cond_30 =( df["age"] >= 30) & (df["age"] < 40 )
df_20 = df[cond_20]
df_30 = df[cond_30]


```

```python
a = df_20["views"].mean()
```

```python
b = df_30["views"].mean()
```

```python
round(abs(b-a), 2) 
```
