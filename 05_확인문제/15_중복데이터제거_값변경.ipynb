```python
# 2. "f3" 컬럼의 값 변환: 결측치는 0, 'silver'는 1, 'gold'는 2, 'vip'는 3

change = {
    "silver" : 1, 
    "gold" : 2, 
    "vip" : 3
}

df["f3"].replace(change,inplace=True)

```
