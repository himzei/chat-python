```python
import pandas as pd 

# 데이터 불러오기
df = pd.read_csv("type1_data2.csv", index_col=0)

# 1. index "2001" 데이터(행)에서 평균보다 큰 값의 개수를 구하시오 
# 2. index "2003" 데이터(행)에서 평균보다 작은 값의 개수를 구하시오 
# 두 개수를 더하시오. 
```

```python
df
```

```python
df = df.T
df
```

```python
m1 = df[2001].mean()
sum(df[2001] > m1)
```

```python
m2 = df[2003].mean()
sum(df[2003] < m2)
```

```python
# 2001 행의 평균
mean_2001 = df.loc[2001].mean()
```

```python
cond1 = df.loc[2001] > mean_2001 
result1 = sum(cond1) 
result1
```

```python
# . index "2003" 데이터(행)에서 평균보다 작은 값의 개수를 구하시오 

mean_2003 = df.loc[2003].mean()
cond2 = df.loc[2003] < mean_2003
result2 = sum(cond2)
result2
```
